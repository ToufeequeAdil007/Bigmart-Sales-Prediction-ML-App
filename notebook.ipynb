# BigMart Advanced Project

This notebook explains a real ML pipeline step by step.

## Step 1: Import Libraries
We load tools for data and ML.

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

## Step 2: Load Data
We read dataset.

```
df = pd.read_csv('Train.csv')
df.head()
```

## Step 3: Feature Engineering
We create better features.

```
df['Price_Per_Visibility'] = df['Item_MRP'] / (df['Item_Visibility']+0.01)
```

## Step 4: Model Training
We train a RandomForest model.

```
X = df[['Item_MRP','Item_Visibility']].dropna()
y = df['Item_Outlet_Sales'].loc[X.index]

model = RandomForestRegressor()
model.fit(X,y)
```

## Step 5: Save Model
We save model for app use.

```
import pickle
pickle.dump(model, open('model.pkl','wb'))
```
